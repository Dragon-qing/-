```python
import matplotlib.pyplot as plt
import random
```

```python
# 创建画布
plt.figure(figsize=(20, 8), dpi=1000)

# 图像绘制
x = [1, 2, 3, 4, 5, 6]
y = [3, 6, 3, 6, 7, 8]
plt.plot(x, y)
# 图像展示
plt.show()
```

```python
x=range(60)
y = [random.uniform(10, 15) for i in x]
```

```python
plt.figure(figsize=(20, 8), dpi = 100)
y_shanghai = [random.uniform(20, 25) for i in x]
plt.plot(x, y_shanghai, label="上海", linestyle="--", color = "r")
plt.plot(x,y, label="北京")

y_ticks = range(40)
x_ticks = ["11min {}sec".format(i) for i in x]

plt.xticks(x[::5],x_ticks[::5])
plt.yticks(y_ticks[::5])
# 添加网格
plt.grid(True, linestyle = "--", alpha = 0.7)
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
# 添加描述
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("一小时温度变化",fontsize = 20)
# 显示图例
plt.legend()
plt.show()
```
