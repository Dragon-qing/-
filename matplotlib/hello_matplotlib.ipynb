```python
import matplotlib.pyplot as plt
import random
```

```python
# 
plt.figure(figsize=(20, 8), dpi=1000)

# 
x = [1, 2, 3, 4, 5, 6]
y = [3, 6, 3, 6, 7, 8]
plt.plot(x, y)
# 
plt.show()
```

```python
x=range(60)
y = [random.uniform(10, 15) for i in x]
```

```python
plt.figure(figsize=(20, 8), dpi = 100)
y_shanghai = [random.uniform(20, 25) for i in x]
plt.plot(x, y_shanghai, label="", linestyle="--", color = "r")
plt.plot(x,y, label="")

y_ticks = range(40)
x_ticks = ["11min {}sec".format(i) for i in x]

plt.xticks(x[::5],x_ticks[::5])
plt.yticks(y_ticks[::5])
# 
plt.grid(True, linestyle = "--", alpha = 0.7)
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
# 
plt.xlabel("")
plt.ylabel("")
plt.title("",fontsize = 20)
# 
plt.legend()
plt.show()
```
