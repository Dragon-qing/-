# 数据离散化处理

```python
data_p = data["p_change"]
```

```python
qcut_r = pd.qcut(data_p, q=10)
```

```python
qcut_r.value_counts()
```

```python
bins = [-100, -7, -5, -3, 0, 3, 5, 7, 100]
```

```python
cut_r = pd.cut(data_p, bins=bins)
```

```python
cut_r.value_counts()
```

```python
data_dummies = pd.get_dummies(cut_r, prefix = "change_")
```

# 合并

```python
data.head()
```

```python
pd.concat([data, data_dummies], axis=1)
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                        'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                        'key2': ['K0', 'K0', 'K0', 'K0'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']})
```

# 交叉表与透视表

```python
data.index
```

```python
time = pd.to_datetime(data.index)
```

```python
time.day
```

```python
time.isocalendar().week
```

```python
data["week"] = time.weekday
```

```python
data
```

```python
data["p_n"] = np.where(data["p_change"]>0, 1, 0)
```

```python
count = pd.crosstab(data["week"], data["p_n"])
```

```python
sum = count.sum(axis = 1)
```

```python
sum
```

```python
per = count.div(sum, axis=0)
```

```python
per
```

```python
per.plot(kind = "bar", stacked = True)
```

```python
data.pivot_table(["p_n"], index="week").plot(kind="bar")
```

# 分组与聚合

```python
col =pd.DataFrame({'color': ['white','red','green','red','green'], 'object': ['pen','pencil','pencil','ashtray','pen'],'price1':[5.56,4.20,1.30,0.56,2.75],'price2':[4.75,4.12,1.60,0.75,3.15]})
```

```python
col
```

```python
col.groupby(["color"])["price1"].mean()
```

```python
col["price1"].groupby(col["color"]).mean()
```

```python
col.groupby(["color"], as_index=False)["price1"].mean()
```
